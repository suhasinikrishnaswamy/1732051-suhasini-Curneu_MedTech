```python
import pandas as pd 
```

```python
u1 = pd.read_csv("User1data.csv")
u2 = pd.read_csv("User2data.csv")
u3 = pd.read_csv("User3data.csv")
u4 = pd.read_csv("User4data.csv")
u5 = pd.read_csv("User5data.csv")
```

```python
u1['Selected'] = u1['Selected'].str.replace("NaN",'0')
u1['Recommended'] = u1['Recommended'].str.replace('(','')
u1['Recommended'] = u1['Recommended'].str.replace(')','')
```

```python
u1['Recommended'] = u1['Recommended'].str[1:]
u1['Selected'] = u1['Selected'].str[1:]
```

```python
y = []
q=0
for val in range(len(u1)) :
    data = u1.iloc[val,4]
    z = []
    for element in data: 
         if str(element) == '-':
                q=1;
         
         if str(element).isupper() == True and q == 0:
                z.append(element)
                a = 1
         
         if (str(element).isdigit() == True or str(element).islower() == True) and a == 1 :
                z.append(element)
         
         if str(element).isupper() == True and q == 1:
                q=0
            
    z=''.join(z)
    a=0
    y.append(z)   
```

```python
u1['Recommended'] = y
```

```python
u1['Selected'] = u1['Selected'].astype(str)
```

```python
y = []
q=0
for val in range(len(u1)) :
    data = u1.iloc[val,5]
    z = []
    for element in data: 
         if str(element) == '-':
                q=1;
         
         if str(element).isupper() == True and q == 0:
                z.append(element)
                a = 1
         
         if (str(element).isdigit() == True or str(element).islower() == True) and a == 1 :
                z.append(element)
         
         if str(element).isupper() == True and q == 1:
                q=0
            
    z=''.join(z)
    a=0
    y.append(z) 
```

```python
u1['Selected'] = y
```
